## Demonstrating the usage of DataTypes and Attributes in Numpy and the usage of Pandas.

```python
a1 = np.array([1, 2, 3])

a2 = np.array([[1, 2, 3],
              [4.5, 5.5, 6.5]])

a3 = np.array([[[1, 2, 3],
                [4, 5, 6],
                [13, 14, 15],
                [13, 14, 15],
                [7, 8, 9]],
                [[10, 11, 12],
                 [13, 14, 15],
                 [22, 24, 55],
                 [13, 14, 15],
                 [16, 17, 18]],
                 [['c', 'v', 'n'],
                  ['a', 's', 'd'],
                  ['z', 'h', 'l'],
                  ['z', 'h', 'l'],
                  ['z', 'h', 'l']]])
```

```python
a1.shape, a1.ndim, a1.dtype, a1.size, type(a1)
```

```python
# What can be noticed in this case is that due to the floating numbers in the array the type of the array is changed from int32 to float64.

a2.shape, a2.ndim, a2.dtype, a2.size, type(a2)
```

```python
# The same case as a2 can be observed here however since there is a string in the array 
# instead of float the type is changed to (U11 as U stands for Unicode and 11 for the length of the array.)

a3.shape, a3.ndim, a3.dtype, a3.size, type(a3)
```

```python
# Taking only the first number of each vector.

a2[:, 0]
```

```python
# Using the Numpy arrays with the Pandas library.

df = pd.DataFrame(np.random.randint(10, size=(3, 3)),
                  columns=['x', 'y', 'z'])
df
```

```python
# Demonstating the usage of already defined Numpy array into a Pandas DataFrame.

df2 = pd.DataFrame(a2)
df2
```

## Demonstration of the usage of Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Creating a simple plot

plt.plot()
```

```python
# Adding data to the plot.

x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x, y)
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")
```

```python
# Using the most common type of plots

import numpy as np

# Line Chart

x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Scatter Chart

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Bar Chart

nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
```
